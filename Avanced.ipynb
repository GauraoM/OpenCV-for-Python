#### Blurring

```python
img = cv.imread(r"C:\Users\gaura\Desktop\Opencv\flower.jpeg") #reading image
cv.imshow('Flower',img) # Displaying the image

# median filter
median = cv.medianBlur(img, 3)
cv.imshow("Median Blur", median)

# bilteral filter(img, d, sigmacolor, sigmaspace)
bilteral = cv.bilateralFilter(img, 3, 15, 15)
cv.imshow("Bilteral Blur", bilteral)

cv.waitKey(0)
```

#### Bitwise Operations

```python
# Creating blank image
blank = np.zeros((400,400), dtype='uint8')

# Drawing reactangle
rectangle = cv.rectangle(blank.copy(), (30,30), (370,370), 255, -1)
# Drawing Circle
circle = cv.circle(blank.copy(), (200,200), 200, 255, -1)

cv.imshow("Rectangle",rectangle)
cv.imshow("circle",circle)

# Bitwise AND
bitwise_and = cv.bitwise_and(reactangle,circle)
cv.imshow('Bitwise AND', bitwise_and)

#### Bitwise OR(superimposed image)
bitwise_or = cv.bitwise_or(reactangle,circle)
cv.imshow('Bitwise OR', bitwise_or)

#### Bitwise OR(non intersecting image)
bitwise_xor = cv.bitwise_xor(reactangle,circle)
cv.imshow('Bitwise XOR', bitwise_xor)

cv.waitKey(0)
```

#### Masking

```python
img = cv.imread(r"C:\Users\gaura\Desktop\Opencv\flower.jpeg") #reading image
cv.imshow('Flower',img) # Displaying the image

# Creating blank image
blank = np.zeros(img.shape[:2], dtype='uint8')

# Drawing a circle
mask = cv.circle(blank, (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)
cv.imshow("Mask",mask)

# Masking the image with and operation
masked = cv.bitwise_and(img,img,mask=mask)
cv.imshow("Masked Image", masked)

cv.waitKey(0)
```

#### Histogram Computation

##### GrayScale Histogram

```python
img = cv.imread(r"C:\Users\gaura\Desktop\Opencv\flower.jpeg") #reading image
cv.imshow('Flower',img) # Displaying the image

# Converting BGR to grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow("Gray", gray)

# GrayScale Histogram
gray_hist = cv.calcHist([gray], [0], None, [256], [0,256])

#plot it
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel('Bins')
plt.ylabel('No. of pixels')
plt.plot(gray_hist)

cv.waitKey(0)
```

##### Color Histogram

```python
img = cv.imread(r"C:\Users\gaura\Desktop\Opencv\flower.jpeg") #reading image
cv.imshow('Flower',img) # Displaying the image

plt.figure()
plt.title("Histogram")
plt.xlabel('Bins')
plt.ylabel('No. of pixels')
colors= ('b','g','r')
# Iterating over colors
for i, col in enumerate(colors):
    hist = cv.calcHist([img], [i], None, [256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])
plt.show()  

cv.waitKey(0)
```

#### Masked Histogram

```python
img = cv.imread(r"C:\Users\gaura\Desktop\Opencv\flower.jpeg") #reading image
cv.imshow('Flower',img) # Displaying the image

blank = np.zeros(img.shape[:2], dtype='uint8')

#Drawing circle
mask = cv.circle(blank, (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)

# Masking the image
masked = cv.bitwise_and(img,img, mask=mask)
cv.imshow('Masked', masked)


plt.figure()
plt.title("Histogram")
plt.xlabel('Bins')
plt.ylabel('No. of pixels')
colors= ('b','g','r')
# Iterating over colors
for i, col in enumerate(colors):
    hist = cv.calcHist([img], [i], mask, [256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])
plt.show() 

cv.waitKey(0)
```

```python
img = cv.imread(r"C:\Users\gaura\Desktop\Opencv\flower.jpeg") #reading image
cv.imshow('Flower',img) # Displaying the image

# Converting BGR to grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow("Gray", gray)

# Simple Thresholding
threshold, thresh = cv.threshold(gray, 150, 255, cv.THRESH_BINARY)
cv.imshow("Simple Thresholded", thresh)

# Inverse Thresholding
threshold, thresh_inv = cv.threshold(gray, 150, 255, cv.THRESH_BINARY_INV)
cv.imshow("Inverse Thresholded", thresh_inv)

# Adaptive Threshold
adaptive_thresh = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 3)
cv.imshow("Adaptive Thresholding",adaptive_thresh )

cv.waitKey(0)
```
